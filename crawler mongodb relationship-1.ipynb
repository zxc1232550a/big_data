```python
import requests
from lxml import etree
import re
import pymongo
import time
import pandas as pd
```

```python
client = pymongo.MongoClient('localhost', 27017)
mydb = client['Dcard']
dcard = mydb['Dcard-1']
global rjson
global data
global last
last =0
```

```python
#定義爬蟲函示
def Crawl(ID):
        link = 'https://www.dcard.tw/_api/posts/' + str(ID)
        r = requests.get(link)
        rjson = r.json()
        data = {'id':rjson['id'],
                'title':rjson['title'],
                'content':rjson['content'],
                'createdAt':rjson['createdAt'],
                'updatedAt':rjson['updatedAt'],
                'commentCount':rjson['commentCount'],
                'forumName':rjson['forumName'],
                'forumAlias':rjson['forumAlias'],
                'gender':rjson['gender'],
                'likeCount':rjson['likeCount'],
                'reactions':rjson['reactions'],
                'topics':rjson['topics']}
        if dcard.find_one({'id': data['id']})is None:      
            dcard.insert_one(data)
            print("1")
        elif dcard.find_one({'id': data['id']})['id'] == data['id']:
            myquery = { "id": data['id'] }
            newvalues = { "$set": { "likeCount": data["likeCount"] } }
            dcard.update_one(myquery, newvalues)
            print("2")
        else:
            dcard.insert_one(data)
            print("3")
        return()
```

```python
url = 'https://www.dcard.tw/_api/forums/relationship/posts?&limit=5'
r = requests.get(url)
rjson = r.json()
Crawl(rjson[0]['id'])
temp = rjson[len(rjson)-1]['id']
print(temp)
for j in range(1000):
    last = str(temp) # 找出爬出資料的最後一筆ID
    url = 'https://www.dcard.tw/_api/forums/relationship/posts?&limit=5&before=' + last
    r = requests.get(url)
    rjson = r.json()
    for i in range(len(rjson)):
        Crawl(rjson[i]['id'])
    temp = rjson[len(rjson)-1]['id']
    print(temp)
    time.sleep(1)
```
